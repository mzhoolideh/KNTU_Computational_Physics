```python
def bisection_method(f, a, b, tolerance):
    # Check that the function has opposite signs at the endpoints
    if f(a) * f(b) >= 0:
        print("The function does not have opposite signs at the endpoints.")
        return None
    
    # Begin the iterative process
    while (b - a) / 2 > tolerance:
        midpoint = (a + b) / 2  # Calculate midpoint
        if f(midpoint) == 0:    # If midpoint is a root
            return midpoint
        elif f(a) * f(midpoint) < 0:  # Root is in [a, midpoint]
            b = midpoint
        else:                          # Root is in [midpoint, b]
            a = midpoint
    
    # Return the midpoint as the approximate root
    return (a + b) / 2

# Test function
f = lambda x: x**2 - 4  # Root at x = ±2
root = bisection_method(f, 1, 3, 1e-5)
print("Approximate root:", root)
```

The function checks if f(a) and f(b) have opposite signs to confirm there's a root within [a, b].
It iteratively calculates the midpoint, checks if the midpoint is a root, and adjusts the interval based on the sign of f(midpoint).
The loop continues until the interval width is less than the tolerance level, and the approximate root is returned.

```python
def factorial(n):
    # Base case: factorial of 0 is 1
    if n == 0:
        return 1
    # Recursive case
    else:
        return n * factorial(n - 1)

# Test the function
print(factorial(5))  # Expected output: 120
print(factorial(0))  # Expected output: 1
```

```python
def fibonacci(n):
    # Base cases
    if n == 0:
        return 0
    elif n == 1:
        return 1
    # Recursive case
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Test the function
print(fibonacci(5))   # Expected output: 5
print(fibonacci(10))  # Expected output: 55
```

```python
def count_characters(s):
    # Initialize an empty dictionary
    char_count = {}
    for char in s:
        # Increment character count if it exists, else set to 1
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    return char_count

# Test cases
print(count_characters("hello"))         # Expected output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}
print(count_characters("mississippi"))   # Expected output: {'m': 1, 'i': 4, 's': 4, 'p': 2}
```

```python
# Define matrices A and B
A = np.arange(1, 10).reshape(3, 3)
B = np.arange(9, 0, -1).reshape(3, 3)

# Matrix multiplication of A and B
C = np.dot(A, B)
print("Matrix C (A * B):\n", C)

# Transpose of the resulting matrix
C_transpose = C.T
print("Transpose of C:\n", C_transpose)

# Inverse of the resulting matrix (if invertible)
try:
    C_inverse = np.linalg.inv(C)
    print("Inverse of C:\n", C_inverse)
except np.linalg.LinAlgError:
    print("Matrix C is not invertible.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for each plot
x = np.linspace(0, 10, 100)
y = x ** 2
categories = ["A", "B", "C", "D"]
values = [5, 3, 8, 6]
uniform_data = np.random.rand(500)
pie_values = [25, 35, 25, 15]
pie_labels = ["Category 1", "Category 2", "Category 3", "Category 4"]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y, color="blue")
axs[0, 0].set_title("Line Plot: y = x^2")

# Bar plot
axs[0, 1].bar(categories, values, color="orange")
axs[0, 1].set_title("Bar Plot")

# Histogram
axs[1, 0].hist(uniform_data, bins=20, color="green")
axs[1, 0].set_title("Histogram of Uniform Data")

# Pie chart
axs[1, 1].pie(pie_values, labels=pie_labels, autopct="%1.1f%%")
axs[1, 1].set_title("Pie Chart")

# Show plot
plt.tight_layout()
plt.show()
```
